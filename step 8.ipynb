```python
df = pd.read_excel("creditcard.xlsx", engine="openpyxl")

print(df.shape)
df.head()
```

```python
df["Amount_log"] = np.log1p(df["Amount"])

df["Hour"] = ((df["Time"] % 86400)//3600).astype(int)

df.drop(["Time","Amount"], axis=1, inplace=True)
```

```python
X = df.drop("Class", axis=1)

y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y
)
```

# Credit Card Fraud Detection Using Supervised Learning

## Summary Report

### Business Problem

Digital payment platforms such as Paytm process millions of transactions every day, making them attractive targets for fraudulent activities. Even a small percentage of fraudulent transactions can result in significant financial losses and negatively affect customer trust. The objective of this project is to build a supervised machine learning model capable of accurately identifying fraudulent credit card transactions while minimizing false alarms. One of the biggest challenges in fraud detection is the highly imbalanced nature of the dataset, where legitimate transactions greatly outnumber fraudulent ones. Because of this imbalance, traditional evaluation metrics such as accuracy are misleading. For example, a model that predicts every transaction as legitimate could achieve over 99% accuracy while completely failing to detect fraud. Therefore, Precision, Recall, F1-Score, and Precision-Recall Area Under Curve (PR-AUC) were used as the primary evaluation metrics since they better measure performance on the minority fraud class.

### Imbalance Handling Strategy

To address the class imbalance problem, three different approaches were evaluated: the original imbalanced dataset, SMOTE (Synthetic Minority Over-sampling Technique), and Random Under Sampling. Logistic Regression models were trained using each strategy and compared using Precision, Recall, F1-Score, and PR-AUC. Based on the evaluation results, the most effective imbalance handling strategy was selected for the final models. SMOTE proved to be the preferred technique because it generates synthetic fraud samples instead of simply duplicating existing ones, allowing the model to learn meaningful fraud patterns while preserving valuable information from legitimate transactions. This approach produced a better balance between fraud detection performance and false positive control compared with the other methods.

### Recommended Model and Threshold

Three supervised learning models were evaluated during this project: Logistic Regression, Random Forest, and XGBoost. Hyperparameter tuning was performed on XGBoost using RandomizedSearchCV with PR-AUC as the optimization metric. Since the default prediction threshold of 0.50 is not always optimal for fraud detection, threshold optimization was also performed using the Precision-Recall curve. Two operating thresholds were analyzed: the threshold that maximized the F1-Score and the threshold that achieved Recall greater than or equal to 90% while maintaining the highest possible Precision. The tuned XGBoost model combined with the F1-optimal threshold demonstrated the strongest overall performance, achieving the best balance between identifying fraudulent transactions and minimizing unnecessary investigations. Therefore, this configuration is recommended for deployment.
